## Question 2: Pointers and References (1 point)

Examine the following Python code carefully:

```python
def check_array(input):
    for idx in range(len(input)):
        if input[idx][0] == "1":
            input[idx] = None
    return input

original_array = ["1_3", "5_2"]
new_array = check_array(original_array)

print(original_array)
print(new_array)
```

What does this code output? Explain **why** it produces that output, using the concepts of pointers and references in Python.

````python
# You may use this cell to experiment with the code above.
````

`check_array` doesn’t create a new list, it directly modifies the same one. That’s because in Python, when you pass a list to a function, you’re passing the reference, not a copy.
So when the loop sets the element starting with `"1"` to `None`, it changes the original list as well.

As a result, both `original_array` and the returned value point to the same list:

```python
[None, '5_2']
[None, '5_2']
```

## Question 3: Algorithm Analysis (2 points)

Consider the following algorithm, which finds the contiguous subarray with the largest sum:

```python
max_sum = None
for i in range(len(array)):
    sum_subarray = 0
    for j in range(i, len(array)):
        sum_subarray += array[j]
        if max_sum is None or max_sum < sum_subarray:
            max_sum = sum_subarray
print(max_sum)
```

Analyze the efficiency of this algorithm:

1. Is this algorithm polynomial? Justify your answer.
2. What is the runtime of this algorithm in **Big-O** notation? Show your reasoning.

````python
# You may use this cell to experiment with the algorithm above.
````

1. **Is it polynomial?**

Yes, the algorithm has two nested loops over the input of size `n = len(array)`. Inside the inner loop, it only does simple constant-time operations like addition, comparison, and assignment.
Because of the nested loops, the overall runtime grows on the order of `n^2`, so it runs in polynomial time.

2. **Big-O runtime**

The outer loop runs `n` times.
For each `i`, the inner loop runs from `j = i` to `n-1`, i.e. `n - i` times.

Total iterations of inner body:

\[
\sum_{i=0}^{n-1} (n-i) = n + (n-1) + \dots + 1 = \frac{n(n+1)}{2}
\]

So the runtime grows on the order of `n^2`, therefore:

- **Time complexity: `O(n^2)`**
- (More precisely, `Theta(n^2)`)
